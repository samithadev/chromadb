```python
# create chroma db client

import chromadb

client = chromadb.Client()
```

```python
# create chromadb collection

collection = client.create_collection("my_collection")
```

```python
collection.add(
    documents= [cricket_news, election_news, football_news, blockchain_news, ai_revolution_news],
    metadatas= [{"source":"cricket"}, {"source":"election"}, {"source":"football"}, {"source":"blockchain"}, {"source":"ai revolution"}],
    ids= ["id1", "id2", "id3", "id4", "id5"]
)
```

```python
# similarity search

results = collection.query(
    query_texts=["technology"],
    n_results= 2
)

results
```

```python
# add data

gaming_news = """
2024 is set to be another fantastic year for video games. For example, Nintendo is releasing a brand-new Princess Peach game; Dragon's Dogma is finally getting a sequel; the next part of the acclaimed Final Fantasy 7 remake is coming out; Star Wars is getting another major release; and that just scratches the surface.
We should expect many more 2024 video game release dates as more presentations happen, like the next State of Play, Nintendo Direct, and all the other online streams.
So, for the gamers who would like to plan their gaming sessions ahead of time, here's a list of the major video games coming out in 2024. We'll try to keep this article updated every week or so with any changes, such as delays, cancellations, and new announcements from the big publishers.
"""

collection.add(
    documents=[gaming_news],
    metadatas=[{"source":"gaming"}],
    ids=["id6"]
)
```

```python
collection.count()
```

```python
# read data

result = collection.get(ids=["id6"])
result
```

```python
# update data

collection.update(
    ids=["id6"],
    documents=["This is a document of gaming"]
)
```

```python
from sentence_transformers import SentenceTransformer
embedding_model = SentenceTransformer("all-mpnet-base-v2")
```

```python
embedding_model.encode(["hello world"])
```

```python
embeddings = embedding_model.encode([cricket_news, football_news])
```

```python
new_collection = client.create_collection("my_new_collection")
```

```python
new_collection.add(
    documents= [cricket_news, football_news],
    embeddings= embeddings,
    metadatas= [{"source":"cricket"},{"source":"football"}],
    ids= ["id1", "id2"]
)
```

```python
new_collection.get()
```

```python
new_collection.query(
    query_embeddings= embedding_model.encode(["test worldcup"]),
    n_results= 1
)
```
